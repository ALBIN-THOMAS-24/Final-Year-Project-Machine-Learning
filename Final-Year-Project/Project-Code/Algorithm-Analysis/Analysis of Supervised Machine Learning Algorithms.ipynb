# Project: Optimal Career Path Finder

**Choosing the right career is not an easy task in anyone’s life. Peer pressure, pressure from parents, personal interests, can conflict the mind and hence, the selection process can be quite frustrating. As a result, people end up choosing wrong careers and finally, end up in jobs that they don’t love doing. So, to help the confused, this project helps them to choose the right career, the digital way. This project examines the various career paths chosen by a diverse class of people, ranging from working professionals and job seekers and try to predict, with maximal accuracy, which career path will be more suitable for the student, using machine learning algorithms.The machine learning model is built using certain rules that are observed from the data set, and generates a probabilistic output of suitable career options that he/she can apply for. In this way, the Student can choose a career in which there is a high chance that he/she will succeed with the highest degree.**

# Science Dataset: 4.5 Lakh Data



 **1. Importing all the necessary libraries**

```python
import matplotlib.pyplot as plt #Data Visualization
import pandas as pd #Reading CSV File
import time #Calculate Time
import numpy as np  #Array
import seaborn as sns #Data Visualization


#Data Preprocessing and Split the dataset into training and  testing set
from sklearn import preprocessing
from sklearn.model_selection import train_test_split


#Library for Supervised ML Algorithms
from sklearn.naive_bayes import GaussianNB  #Gaussian Naive Bayes
from sklearn import linear_model #Logistic Regression
from sklearn.neural_network import MLPClassifier #MLPClassifier
from sklearn.ensemble import AdaBoostClassifier #Adaboost Classifier
from sklearn.tree import DecisionTreeClassifier #Decision Tree




#Model Evaluation
import sklearn.metrics as metrics #Finding the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix

```

**2. Read the Science-Dataset.csv File**

```python
data = pd.read_csv("C:\\machine\\science-dataset.csv")
data.head(10) #DISPLAY TOP 10 ROWS
```

**3.Data Visualization and Data Scaling**

```python
sns.set()
sns.countplot(x="CAREER",data=data)
plt.show()
print("\n")

#Creating the pie chart
career=['ENGINEER','MEDICINE','OTHERS']
Engineer=len(data.loc[data['CAREER']==0])
Medicine=len(data.loc[data['CAREER']==1])
Others=len(data.loc[data['CAREER']==2])
print("Engineering=",Engineer)
print("Medicine=",Medicine)
print("Others=",Others)
print("\n")
count=np.array([Engineer,Medicine,Others])
plt.pie(count, labels = career)
plt.title('Pie Chart Showing different careers chosen by various people')
#Show the plot
plt.show()
print("\n\n")


print("Correlation Matrix")
df_small = data.iloc[:,:23]
correlation_mat = df_small.corr()
plt.subplots(figsize=(25,10))
sns.heatmap(correlation_mat, annot = True)

plt.show()


#Data Scaling
X=data.drop(['CAREER'], axis = 1).values 
print("Scaled Data")
Standardisation = preprocessing.StandardScaler()#Standard Scaling
AFTER_SCALE= Standardisation.fit_transform(X) 
print(AFTER_SCALE)
print("\n")
X=AFTER_SCALE #Input Features
y = data['CAREER'] #Output Label
print("450000 Rows and 23 Features")



#Split the Data into training and testing set 
#80% data used for training and 20% data used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
print("\n")




```

**4.. Analysis of Supervised Machine Learning Algorithms**

**4.1 : Naive Bayes**

```python

start=time.time() #Start Time

naive=GaussianNB()#Gaussian Naive Bayes
naive.fit(X_train,y_train)#training 
y_pred = naive.predict(X_test) #Predict

end=time.time()#End Time

print("Total Time Taken => "+str(end-start)+str(" ")+str("Seconds"))
print("\n")



#Model Evaluation
print("NAIVE-BAYES-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(naive, X_test, y_test,cmap=plt.cm.Blues) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report





```

**4.2 : Logistic Regression**

```python

start=time.time() #Start Time

log=linear_model.LogisticRegression() #Logistic Regression
log.fit(X_train,y_train) #Training
y_pred=(log.predict(X_test)) #Predict

end=time.time()#End Time

print("Total Time Taken => "+str(end-start)+str(" ")+str("Seconds"))
print("\n")


#Model Evaluation
print("LOGISTIC-REGRESSION-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(log, X_test, y_test,cmap=plt.cm.Blues) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report

```

**4.3: Multilayer Perceptron with 2 Hidden Layers**

```python

start=time.time()#Start Time

#Two Hidden Layers 8 Neurons Each
mlp= MLPClassifier(hidden_layer_sizes=(8,8), activation='relu', solver='adam',max_iter=300) 
mlp.fit(X_train,y_train)#Training
y_pred=(mlp.predict(X_test))#Predict

end=time.time()#End Time

print("Total Time Taken => "+str(end-start)+str(" ")+str("Seconds"))
print("\n")

#Model Evaluation
print("MLP-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(mlp, X_test, y_test,cmap=plt.cm.Blues) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
```

**4.4 : Adaboost Classifier**

```python

start=time.time() #Start Time

abc = AdaBoostClassifier() #Adaboost Classifier
model = abc.fit(X_train, y_train) #Training
y_pred = model.predict(X_test) #Predict

end=time.time()#End Time

print("Total Time Taken => "+str(end-start)+str(" ")+str("Seconds"))
print("\n")

#Model Evaluation
print("Adaboost-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(abc, X_test, y_test,cmap=plt.cm.Blues) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
```

**4.5 : Decision Tree**

```python

Start = time.time () #Start Time

Decision_tree= DecisionTreeClassifier(criterion = 'gini') #Decision Tree
Decision_tree.fit(X_train, y_train)#Training
y_pred = Decision_tree.predict(X_test) #Predict

end=time.time() #End Time

    
print("Total Time Taken => "+str(end-start)+str(" ")+str("Seconds"))
print("\n")

#Model Evaluation
print("Decision Tree-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(Decision_tree, X_test, y_test,cmap=plt.cm.Blues) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
```
