**1. Science dataset : 50000 data**

```python
import pandas as pd
import numpy as np
data = pd.read_csv("C:\\machine\\science-dataset.csv")
data.head(10) #DISPLAY TOP 10 ROWS


#Data Scaling
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
X=data.drop(['CAREER'], axis = 1).values 
X=X[:50000]
print("Scaled Data")
Standardisation = preprocessing.StandardScaler()#Standard Scaling
AFTER_SCALE= Standardisation.fit_transform(X) 
print(AFTER_SCALE)
print("\n")
X=AFTER_SCALE #Input Features
y = data['CAREER'] #Output Label
y=y[:50000]

print(X.shape)
print(y.shape)

#Split the Data into training and testing set 
#80% data used for training and 20% data used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
print("\n")




print("SVM with rbf kernel")
from sklearn.svm import SVC
#Model Evaluation
import sklearn.metrics as metrics #Finding the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix
import matplotlib as plt
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test) #Predict

#Model Evaluation
print("SVM-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(svm, X_test, y_test) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
print("\n\n")



print("SVM with linear kernel")
from sklearn.svm import SVC
#Model Evaluation
import sklearn.metrics as metrics #Finding the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix
import matplotlib as plt
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test) #Predict

#Model Evaluation
print("SVM-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(svm, X_test, y_test) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
print("\n\n")



print("SVM with Polynomial kernel")
from sklearn.svm import SVC
#Model Evaluation
import sklearn.metrics as metrics #Finding the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix
import matplotlib as plt
svm = SVC(kernel='poly')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test) #Predict

#Model Evaluation
print("SVM-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(svm, X_test, y_test) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
print("\n\n")



print("SVM with sigmoid kernel")
from sklearn.svm import SVC
#Model Evaluation
import sklearn.metrics as metrics #Finding the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix
import matplotlib as plt
svm = SVC(kernel='sigmoid')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test) #Predict

#Model Evaluation
print("SVM-ACCURACY => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy
print("\n")
print("Confusion Matrix")
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(svm, X_test, y_test) #Confusion Matrix
print(disp.confusion_matrix)
print("\n")
print("Classification Report")
print(classification_report(y_test,y_pred)) #Classification Report
print("\n\n")
```
