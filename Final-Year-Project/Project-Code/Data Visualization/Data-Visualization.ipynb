# 1. Science Dataset: 4.5 lakh data

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
data = pd.read_csv("C:\\machine\\science-dataset.csv")
```

**1. Barchart**

```python
sns.set()
sns.countplot(x="CAREER",data=data)
print("\n")
```

**2. Scatter Plot**

```python

print("Memory Score vs IQ")
data.plot(kind ="scatter",
          x ='IQ',
          y ='M-SCORE',color='blue')
```

```python
print("OPENNESS vs EXTRAVERSION")
data = pd.read_csv("C:\\machine\\science-dataset.csv")
data.plot(kind ="scatter",
          x ='OPENNESS',
          y ='EXTRAVERSION',color='orange')
```
