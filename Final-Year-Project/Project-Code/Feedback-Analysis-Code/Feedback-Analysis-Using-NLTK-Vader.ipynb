# STUDENT FEEDBACK ANALYSIS USING VADER NLTK

**VADER is a model used for text sentiment analysis that is sensitive to both polarity (positive/negative) and intensity (strength) of emotion. It is available in the NLTK package and can be applied directly to unlabeled text data.**

```python
#Steps
#1. Input Text
#2. Clean the text by removing punctuation,symbols,numbers,etc.
#3. 
    #3.1 Initialize positive_count=0 and negative_count=0
    #3.2 Find the polarity of each word in text 
    #3.3 if(compound of word>=0.05 then increase positive count)
    #3.4 otherwise, increase negative count
    #3.5 Handling the negation 
    #3.6 check if positive_count > negative_count:
        #Display("The text is Positive")
    #3.7 Otherwise, Display("The text is Negative")
```

```python
#STUDENT FEEDBACK ANALYSIS USING VADER NLTK

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vader=SentimentIntensityAnalyzer()

while(1):
    #Input text
    print("\n")
    m=input(" ")


    def feedback_Analysis(data_after_clean):
        print("\n")
        print("After Cleaning the text")
        print(data_after_clean)
        print("\n")

        pos=0
        neg=0
        flag=0
        res = data_after_clean.split()
        for i in res:
            #Negation Handling
            if(i=='not'):
                flag=1
                
            k=vader.polarity_scores(i)
            if(k['compound']>=0.05):
                pos=pos+1
            elif(k['neu']==1 or k['compound']==0):
                continue
            else:
                neg=neg+1

        if(pos>neg and flag==1):
            neg=pos+1
            flag=0
        if(neg>pos and flag==1):
            pos=neg+1
            flag=0
       
        print("Output:")
        if(pos>neg):
            print("The feedback is positive!!")
        else:
            print("The feedback is negative!!")


    def data_cleaning():
        #convert text to lowerCase
        data=m.lower()

        # Remove Punctuations
        punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        data_after_clean=""
        for char in data:
            if char not in punctuations:
                data_after_clean = data_after_clean + char

        feedback_Analysis(data_after_clean)
    data_cleaning()
    
    print("\n\nDo you Want to Contine Press 1? or Press 3 to exit")
    choice=input()
    if(choice=="1"):
        continue
    elif(choice=="3"):
        break
    else:
        break
        
```

```python
feedback=data['Feedback']
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vader=SentimentIntensityAnalyzer()
response=[]
for i in range(4999):
    input_feedback=feedback[i]
    data=input_feedback.lower()
    # Remove Punctuations
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    data_after_clean=""
    for char in data:
        if char not in punctuations:
            data_after_clean = data_after_clean + char
    pos=0
    neg=0
    flag=0
    res = data_after_clean.split()
    for i in res:
        #Negation Handling
        if(i=='not'):
            flag=1
                
        k=vader.polarity_scores(i)
        if(k['compound']>=0.05):
            
            pos=pos+1
        elif(k['neu']==1 or k['compound']==0):
            continue
        else:
            neg=neg+1

    if(pos>neg and flag==1):
        neg=pos+1
        flag=0
    if(neg>pos and flag==1):
        pos=neg+1
        flag=0
       
    if(pos>neg):
        response.append('1')
    else:
         response.append('0')
```

```python
responses=data['Sentiment']
result=[]
for i in range(len(responses)):
    if(responses[i]=='positive'):
        result.append('1')
    else:
        result.append('0')

correct_response=0
maximum=5000
for i in range(4999):
    if(result[i]==response[i]):
        correct_response+=1

print("Accuracy= ",(correct_response/maximum)*100)    
```
