**1. Science dataset**

```python
print("Careers related to Science Stream\n\n")
import pandas as pd
import numpy as np
data = pd.read_csv("C:\\machine\\science-dataset.csv")
data.head(10) #DISPLAY TOP 10 ROWS


def svm_train(i,j):

    #Data Scaling
    from sklearn import preprocessing
    from sklearn.model_selection import train_test_split
    X=data.drop(['CAREER'], axis = 1).values 
    print("Scaled Data")
    Standardisation = preprocessing.StandardScaler()#Standard Scaling
    AFTER_SCALE= Standardisation.fit_transform(X) 
    print(AFTER_SCALE)
    print("\n")
    X=AFTER_SCALE #Input Features
    X=X[i:j]
    y = data['CAREER'] #Output Label
    y=y[i:j]

    print(X.shape)
    print(y.shape)

    #Split the Data into training and testing set 
    #80% data used for training and 20% data used for testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
    print("\n")



    print("SVM with rbf kernel")
    from sklearn.svm import SVC
    #Model Evaluation
    import sklearn.metrics as metrics #Finding the accuracy of the model
    from sklearn.metrics import classification_report,confusion_matrix #classification Report  and Confusion matrix
    import matplotlib as plt
    svm = SVC(kernel='rbf')
    svm.fit(X_train, y_train)

    y_t=svm.predict(X_train)#predict train
    y_pred = svm.predict(X_test) #Predict test

    #Model Evaluation
    print("SVM_training_Accuracy => "+str(metrics.accuracy_score(y_train, y_t)*100)+str(" %"))#Accuracy
    print("SVM_testing_Accuracy => "+str(metrics.accuracy_score(y_test, y_pred)*100)+str(" %"))#Accuracy

    print("\n")
    print("Confusion Matrix")
    from sklearn.metrics import plot_confusion_matrix
    disp = plot_confusion_matrix(svm, X_test, y_test) #Confusion Matrix
    print(disp.confusion_matrix)
    print("\n")
    print("Classification Report")
    print(classification_report(y_test,y_pred)) #Classification Report
    print("\n\n")
    #pickle model
    import pickle
    pickle.dump(svm, open('science_model.pkl','wb'))
    


print("Train and test on 50000 data")
i=0
j=50000
svm_train(i,j)
print('\n')
print("Train and test on next 50000 data")
i=50000
j=100000
svm_train(i,j)
print("\n")
print("Train and test on next 50000 data")
i=100000
j=150000
svm_train(i,j)
print("\n")
print("Train and test on next 50000 data")
i=150000
j=200000
svm_train(i,j)
print("\n\n")
print("Train and test on next 50000 data")
i=200000
j=250000
svm_train(i,j)
print("\n\n")
print("Train and test on next 50000 data")
i=250000
j=300000
svm_train(i,j)
print("\n\n")
print("Train and test on next 50000 data")
i=350000
j=400000
svm_train(i,j)
print("\n\n")
print("Train and test on next 50000 data")
i=400000
j=450000
svm_train(i,j)
print("\n\n")






```
