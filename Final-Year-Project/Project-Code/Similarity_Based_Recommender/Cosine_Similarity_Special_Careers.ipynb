```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
#import nltk
#nltk.download('stopwords')
#nltk.download('punkt')

#from google.colab import drive
#drive.mount('/content/drive')
print("Entered text:")
a=input()

def similarity(a,text):
    
    #convert to lowercase
    a=a.lower()
   
    #tokenize and remove stopwords
    a= word_tokenize(a)
    a= [word for word in a if not word in stopwords.words()]
    a = (" ").join(a)
    
    
    #tokenize and remove stopwords
    text=text.lower()
    text= word_tokenize(text)
    text= [word for word in text if not word in stopwords.words()]
    text = (" ").join(text)
   

    #Snow_ball stemming
    snow_stemmer = SnowballStemmer(language='english')
    a=snow_stemmer.stem(a)
    text=snow_stemmer.stem(text)
   
    
   
    X=a #user_input
    Y=text #Text File
    
    #cosine similarity
    text=[X,Y]
    cv=CountVectorizer()
    count_matrix=cv.fit_transform(text)
    sim_score=cosine_similarity(count_matrix)
    
    #return similarity score
    return sim_score[0][1]

print("\n")
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/CP.txt', encoding="utf8") as file:
    cp= file.read()
print("Commercial Pilot=",similarity(a,cp))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/CW.txt', encoding="utf8") as file:
    cw= file.read()
print("Content Writer=",similarity(a,cw))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/DET.txt', encoding="utf8") as file:
    det= file.read()
print("Detectives =",similarity(a,det))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/DI.txt', encoding="utf8") as file:
    di= file.read()
print("Dance Instructor=",similarity(a,di))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/EH.txt', encoding="utf8") as file:
    eh= file.read()
print("Ethical hacker=",similarity(a,eh))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/ID.txt', encoding="utf8") as file:
    id= file.read()
print("Interior Designer=",similarity(a,id))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/MA.txt', encoding="utf8") as file:
    ma= file.read()
print("Makeup Artist=",similarity(a,ma))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/MOD.txt', encoding="utf8") as file:
    mod= file.read()
print("Modelling=",similarity(a,mod))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/PGR.txt', encoding="utf8") as file:
    pgr= file.read()
print("Photographer=",similarity(a,pgr))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/PT.txt', encoding="utf8") as file:
    pt= file.read()
print("Personal Trainer=",similarity(a,pt))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/PY.txt', encoding="utf8") as file:
    py= file.read()
print("Professional Youtuber=",similarity(a,py))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/REA.txt', encoding="utf8") as file:
    rea= file.read()
print("Real Estate Agent=",similarity(a,rea))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/SMP.txt', encoding="utf8") as file:
    smp= file.read()
print("Stock Market Professional=",similarity(a,smp))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/TC.txt', encoding="utf8") as file:
    tc= file.read()
print("Travel Consultant=",similarity(a,tc))
with open('/content/drive/MyDrive/Project files/Recomender/SPECIAL/WP_EM.txt', encoding="utf8") as file:
    wp= file.read()
print("Wedding Planner=",similarity(a,wp))
```
